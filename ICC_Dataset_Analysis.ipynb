```python
def get_innings_score(innings_deliveries):
  innings_deliveries_df=pd.DataFrame(columns=['delivery','runs_scored'])
  for delivery in innings_deliveries:
    delivery_number=(list(delivery.keys())[0])
    extras=(delivery[delivery_number]['runs']['extras'])
    total=(delivery[delivery_number]['runs']['total'])
    innings_deliveries_df=innings_deliveries_df.append({'delivery':delivery_number,'runs_scored':total},ignore_index=True)
  innings_deliveries_df['over']=innings_deliveries_df['delivery'].apply(lambda s:int(s))
  innings_deliveries_df['over']=innings_deliveries_df['over']+1

  innings_runs_df=innings_deliveries_df.groupby('over').agg({'runs_scored':sum}).reset_index()
  innings_runs_df['total_score']=innings_runs_df['runs_scored'].cumsum()
  return innings_runs_df
```

```python
first_innings_runs_df=get_innings_score(first_innings_deliveries)
second_innings_runs_df=get_innings_score(second_innings_deliveries)
```

```python
import seaborn as sns
sns.lineplot(data=first_innings_runs_df,x='over',y='total_score')
sns.lineplot(data=second_innings_runs_df,x='over',y='total_score')
```
